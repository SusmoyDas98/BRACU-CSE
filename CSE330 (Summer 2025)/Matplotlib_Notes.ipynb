## **Key Terms**

**Data Point**: A single value or coordinate representing an observation in a plot.

**X-axis / Y-axis**: Horizontal and vertical axes used to define the coordinate system of a plot.

**Figure**: The overall container that holds one or more plots, including titles and legends.

**Axes**: The area of the figure where data is plotted; contains the x and y axes.

**Plot**: A visual representation of data using lines, markers, or other symbols.

**Marker**: A symbol used to represent individual data points on a plot.

**Line Style**: The appearance of the plotted lines (e.g., solid, dashed, dotted).

**Color**: The attribute that defines the visual color of plot elements.

**Legend**: A key that explains the symbols, colors, or styles used in the plot.

**Label**: Text assigned to axes or data series to describe them.

**Title**: The heading of the plot that describes its purpose or content.

**Grid:** A structured arrangement of rows and columns.

**Function:** A block of organized, reusable code that performs a single, related action.

**Method:** A function that belongs to an object.

**Parameters:** Variables listed inside the parentheses in a function definition.


**Keyword Arguments:** Arguments preceded by a keyword (e.g., `name=value`) in a function call.


**Object-Oriented API:** An Application Programming Interface designed around objects and their interactions.


**DPI:** Dots Per Inch, a measure of spatial printing or video dot density.


**Backend:** The part of a computer system or application that is not directly accessed by the user, responsible for data storage and processing.

# **Matplotlib Plotting Functions: Syntax and Use (with Optional Arguments)**

This document outlines the syntax and purpose of common Matplotlib plotting functions, as depicted in the provided diagram. All these functions typically belong to the `matplotlib.pyplot` module, commonly imported as `plt`.

---
---

---

## 1. Grid

* **Syntax:**
    ```python
    plt.grid(visible=True, axis='both', linestyle='-', linewidth=0.5, color='gray', alpha=0.7)
    # plt.grid(True) is a common shorthand for visible=True with default styles
    ```
* **Use:** Adds a grid to the plot.
    * `visible` (bool, default: `True` if no arguments, `False` otherwise): Whether to show the grid.
    * `axis` ('x', 'y', or 'both', default: 'both'): Which axes to apply the grid to.
    * `linestyle` (str, default: '-'): The line style of the grid (e.g., '-', '--', '-.', ':').
    * `linewidth` (float, default: 0.8): The line width of the grid.
    * `color` (str, default: 'gray'): The color of the grid lines.
    * `alpha` (float, default: 0.5): The transparency of the grid lines (0.0 fully transparent, 1.0 fully opaque).

---

## 2. Labels

* **Syntax:**
    ```python
    plt.xlabel('Your X-axis Label', fontsize=12, color='blue', loc='center')
    plt.ylabel('Your Y-axis Label', fontsize=12, color='green', rotation='vertical')
    ```
* **Use:** Sets labels for the x and y axes.
    * `label` (str): The text for the label.
    * `fontsize` (float or str): Font size of the label (e.g., `12`, `'large'`).
    * `color` (str): Color of the label text.
    * `loc` (str, for `xlabel` only): Horizontal alignment of the label ('left', 'center', 'right').
    * `rotation` (str or float, for `ylabel` only): Rotation of the label ('horizontal', 'vertical', or angle in degrees).

---

## 3. Plotting

* **Syntax:**
    ```python
    plt.plot(x_values, y_values,
             color='blue',       # Line color
             linestyle='--',     # Line style
             linewidth=2,        # Line width
             marker='o',         # Marker style
             markersize=8,       # Marker size
             label='My Data Series') # Label for the legend
    # Example using format string (shorthand for color, marker, linestyle):
    # plt.plot([1, 2, 3], [4, 5, 6], 'go-', label='Data 1') # green, circle markers, solid line
    ```
* **Use:** The fundamental function for creating 2D plots.
    * `x_values`, `y_values` (array-like): Data points for the plot.
    * `format_string` (str, optional): A shorthand string specifying color, marker, and line style (e.g., `'r--'` for red dashed line, `'bo'` for blue circles).
    * `color` (str): Explicit line color.
    * `linestyle` (str): Explicit line style (e.g., '-', '--', ':', '-.').
    * `linewidth` (float): Line width in points.
    * `marker` (str): Marker style (e.g., 'o', 'x', '^', 's').
    * `markersize` (float): Size of the markers.
    * `label` (str): Text to appear in the legend for this plot element.

---

## 4. Title

* **Syntax:**
    ```python
    plt.title('Your Plot Title', fontsize=14, color='darkred', loc='center')
    ```
* **Use:** Adds a main title to the entire plot.
    * `title` (str): The text for the title.
    * `fontsize` (float or str): Font size of the title.
    * `color` (str): Color of the title text.
    * `loc` (str): Horizontal alignment of the title ('left', 'center', 'right').

---

## 5. Legend

* **Syntax:**
    ```python
    plt.legend(loc='upper left',           # Location of the legend
               fontsize='small',         # Font size of legend text
               frameon=True,             # Whether to draw a frame around the legend
               fancybox=True,            # Rounded box if frameon is True
               shadow=True,              # Whether to draw a shadow behind the legend
               ncol=1,                   # Number of columns for legend entries
               borderpad=0.5,            # Padding between legend border and content
               labelspacing=0.5)         # Vertical space between legend entries
    ```
* **Use:** Displays a legend on the plot, identifying different plot elements by their `label`.
    * `loc` (str or int, default: 'best'): The location of the legend. Common values include:
        * `'best'` (0)
        * `'upper right'` (1)
        * `'upper left'` (2)
        * `'lower left'` (3)
        * `'lower right'` (4)
        * `'right'` (5)
        * `'center left'` (6)
        * `'center right'` (7)
        * `'lower center'` (8)
        * `'upper center'` (9)
        * `'center'` (10)
    * `fontsize` (float or str): Font size of the legend text.
    * `frameon` (bool): Whether to draw a frame around the legend.
    * `fancybox` (bool): Whether to draw a rounded box if `frameon` is `True`.
    * `shadow` (bool): Whether to draw a shadow behind the legend.
    * `ncol` (int, default: 1): The number of columns that the legend has.
    * `borderpad` (float): Padding between the legend border and its contents.
    * `labelspacing` (float): Vertical space between the legend entries.

---

## 6. Axis Ticks

* **Syntax:**
    ```python
    plt.xticks(ticks=[0, 1, 2],
               labels=['Category A', 'Category B', 'Category C'],
               rotation=45,          # Rotate tick labels
               fontsize=10)
    plt.yticks(ticks=[0.0, 0.5, 1.0],
               labels=['Low', 'Medium', 'High'],
               color='purple')
    ```
* **Use:** Used to control the properties of the x and y-axis ticks and their labels.
    * `ticks` (array-like, optional): A list of numerical locations where ticks should be placed. If not provided, Matplotlib generates default locations.
    * `labels` (array-like, optional): A list of strings to use as labels for the ticks. Must be the same length as `ticks`.
    * `rotation` (float or str): Rotation of the tick labels in degrees.
    * `fontsize` (float or str): Font size of the tick labels.
    * `color` (str): Color of the tick labels.

---

## 7. Axis Limits

* **Syntax:**
    ```python
    plt.xlim(xmin=0, xmax=10)
    plt.ylim(ymin=-5, ymax=5)
    # You can also omit one argument to keep the current limit for that side:
    # plt.xlim(xmin=0) # Sets min x-limit, keeps max x-limit unchanged
    ```
* **Use:** Sets the minimum and maximum values for the respective axes.
    * `xmin`, `xmax` (float): The lower and upper bounds for the x-axis.
    * `ymin`, `ymax` (float): The lower and upper bounds for the y-axis.

---

## 8. Display

* **Syntax:**
    ```python
    plt.show(block=True)
    ```
* **Use:** Crucial function to display the generated plot.
    * `block` (bool, default: `True`): Whether to block execution until the plot window is closed. In interactive environments like Colab, this is often `False` by default so your code continues to run. For scripts, `True` typically means the plot window stays open until you close it.

````python
x = [1,2,3,4] # X-AXIS

y = [10,20,15,25] # Y-AXIS

plt.grid(True)

plt.plot(x,y)

plt.show()
````

# **Data Visualization Tools and Their Purposes (with Matplotlib Syntax)**

## **1. Bar Charts**

* **Purpose:**
    * **Category *Comparison*:** Excellent for comparing discrete categories or groups. Each bar represents a category, and its length or height corresponds to a value (e.g., sales by product, population by city).
    * **Data Analysis:** Useful for identifying trends, highs, and lows across different categories. They make it easy to see which categories are larger or smaller than others.

* **Syntax (`plt.bar()` for vertical, `plt.barh()` for horizontal):**
    * **Basic:** `plt.bar(x_categories, y_values)`
    * **With Options (Vertical):**
        `plt.bar(x, height, width=0.8, color='blue', edgecolor='black', linewidth=1, align='center', label='Series Name')`
        `plt.xlabel('X-axis Label')`
        `plt.ylabel('Y-axis Label')`
        `plt.title('Chart Title')`
        `plt.legend()`
        `plt.show()`
    * **With Options (Horizontal):**
        `plt.barh(y, width, height=0.8, color='green', edgecolor='black', xerr=error_values, label='Series Name')`
        `plt.xlabel('X-axis Label')`
        `plt.ylabel('Y-axis Label')`
        `plt.title('Chart Title')`
        `plt.legend()`
        `plt.show()`

---

````python
import matplotlib.pyplot as plt
````

````python
product = ['A',"B","C","D"]

sales = [1000,1500,800,1200]

plt.bar(product, sales, color = 'skyblue', label = 'Sales 2025')

plt.xlabel("Product")

plt.ylabel("Sales")

plt.title("Product Sales Comparison", fontsize = 20)

plt.legend()

plt.show()
````

````python

# plt.barh() FOR HORIZONTAL BARS

plt.barh(product, sales, color = 'skyblue', label = 'Sales 2025')

plt.xlabel("Sales")

plt.ylabel("Product")

plt.title("Product Sales Comparison", fontsize = 20)

plt.legend()

plt.show()
````

## **2. Pie Charts**

* **Purpose:**
    * **Proportion Illustration:** Primarily used to show the proportion or percentage of different categories within a whole. Each slice represents a category, and the size of the slice is proportional to its value.
    * **Whole Representation:** Ideal for representing parts of a whole, where all segments add up to 100%. They quickly convey the relative contribution of each category to the total.

* **Syntax (`plt.pie()`):**
    * **Basic:** `plt.pie(sizes)`
    * **With Options:**
        `plt.pie(sizes, explode=explode_tuple, labels=category_labels, colors=color_list, autopct='%1.1f%%', shadow=True, startangle=90, pctdistance=0.8, labeldistance=1.1, wedgeprops={'edgecolor': 'black', 'linewidth': 1})`
        `plt.axis('equal')` (Ensures circle shape)
        `plt.title('Chart Title')`
        `plt.show()`

* ***autopct='%1.1f%%*** --> Shows the percentage coming from each slice.
    * **f** means the format
    * ***1.1*** means 1 digit before the decimal and 1 digit after the decimal
    * ***%%*** means the ***%*** sign must be invloved in it.

---

````python
import matplotlib.pyplot as plt
````

````python
regions = ["North", "South", "East", "West"]

revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels = regions, autopct = '%1.1f%%', colors = ['Gold', 'SKyblue','Lightgreen','Coral'])

plt.title("Revenue Contribution By Region")

plt.show()
````

## **3. Histograms**

* **Purpose:**
    * **Numerical Distribution:** Used to visualize the distribution of a continuous numerical variable. They divide the data into "bins" and count how many data points fall into each bin, displaying these counts as bars.

* **Syntax (`plt.hist()`):**

        `plt.xlabel('X-axis Label')`
        `plt.ylabel('Y-axis Label')`
        `plt.title('Chart Title')`

        `plt.show()`

## **4. Scatter Plots**

* **Purpose:**
    * **Relationship Visualization:** Primarily used to display the relationship between two continuous numerical variables. Each point on the plot represents an observation, with its position determined by the values of the two variables.
    * **Pattern Identification:** Excellent for identifying trends, clusters, outliers, and correlations between variables. You can easily see if there's a positive, negative, or no correlation, and if the relationship is linear or non-linear.

* **Syntax (`plt.scatter()`):**
    * **Basic:** `plt.scatter(x_values, y_values)`

        `plt.scatter(x, y, s=size_of_markers, c=colors_of_markers, marker='o', alpha=0.7, edgecolors='black', linewidths=0.5, label='Data Points')`

## **1. Grid Layout (`plt.subplot()`)**

* **Purpose:**
    * **Quick Creation:** Provides a fast way to create a single subplot in a grid.
    * **Positional Control:** Allows specifying the exact row, column, and index for the current subplot.

* **Syntax (`plt.subplot()`):**
    * **Basic:** `plt.subplot(nrows, ncols, index)` or `plt.subplot(NMN)`
    * **Example Usage Pattern:**
        ```python
        import matplotlib.pyplot as plt
        # Assuming x1, y1, x2, y2 are defined
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
        plt.plot(x1, y1)
        plt.title('Subplot 1')

        plt.subplot(122) # Shorthand for 1 row, 2 columns, 2nd plot
        plt.plot(x2, y2)
        plt.title('Subplot 2')
        plt.show()
        ```
    * `nrows` (int): Number of rows in the grid.
    * `ncols` (int): Number of columns in the grid.
    * `index` (int): The index of the current subplot (starting from 1, reading across rows then down).
    * `NMN` (int): Shorthand form like `121` for `(nrows=1, ncols=2, index=1)`.

---

````python
import matplotlib.pyplot as plt
````

````python
x = [1,2,3,4]

y = [10,20,15,25]

plt.subplot(1,2,1)

plt.title("Line Chart")

plt.plot(x,y)

plt.subplot(1,2,2)

plt.title("Bar Chart")

plt.bar(x,y)


plt.show()
````

## **2. Object-Oriented Creation (`plt.subplots()`)**

* **Purpose:**
    * **Preferred Method:** The recommended and more flexible way to create a figure and a set of subplots.
    * **Object-Oriented Control:** Returns `Figure` and `Axes` objects, allowing fine-grained control over each subplot using their methods (e.g., `ax.set_title()`, `ax.set_xlabel()`).

* **Syntax (`plt.subplots()`):**
    * **Basic:** `fig, ax = plt.subplots()` (creates a single plot)
    * **Multiple Subplots:** `fig, axes = plt.subplots(nrows=2, ncols=2)`
    * **With Options:**
        ```python
        import matplotlib.pyplot as plt
        # Assuming data for plotting is defined
        fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 6), sharex=True, sharey=False, squeeze=True)
        # axes will be a 1D array if squeeze=True and either nrows or ncols is 1

        axes[0].plot(data1_x, data1_y)
        axes[0].set_title('Top Subplot')
        axes[0].set_ylabel('Value A')

        axes[1].plot(data2_x, data2_y, color='red')
        axes[1].set_title('Bottom Subplot')
        axes[1].set_xlabel('Common X-axis')
        axes[1].set_ylabel('Value B')

        fig.suptitle('Overall Figure Title', fontsize=16) # Figure-wide title
        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make space for suptitle
        plt.show()
        ```
    * `nrows` (int, default: 1): Number of rows in the grid.
    * `ncols` (int, default: 1): Number of columns in the grid.
    * `figsize` (tuple): Width and height of the figure in inches (e.g., `(8, 6)`).
    * `sharex` (bool or `Axes`): If `True`, all subplots share the same x-axis limits and ticks. If an `Axes` object, it shares with that specific axes.
    * `sharey` (bool or `Axes`): Similar to `sharex` but for the y-axis.
    * `squeeze` (bool, default: `True`): If `True`, the returned `axes` array is squeezed to a 1D array if possible (e.g., for `nrows=1, ncols=2`, `axes` will be `[ax0, ax1]` instead of `[[ax0, ax1]]`).
    * **Returns:** `fig` (a `Figure` object), `ax` or `axes` (an `Axes` object for a single plot, or an array of `Axes` objects for multiple plots).

---


````python
fig, ax = plt.subplots(1, 2, figsize = (10,5))

x  = [1,2,3,4]

y = [10,20,15,25]

ax[0].plot(x,y , color = 'red')

ax[0].set_title("Line Plot")

ax[1].bar(x,y, color = 'lightgreen', edgecolor = "black")

ax[1].set_title("Bar Chart")

plt.tight_layout()

fig.suptitle("Comparison of Line and Bar Chart")

plt.show()
````

## **Saving Visualizations (`plt.savefig()`)**

* **Purpose:**
    * **Persistence:** Saves the currently active Matplotlib figure to a file on your disk.
    * **Sharing:** Enables easy sharing of visualizations with others (e.g., for collaboration or publication).
    * **Future Use:** Allows plots to be reused for analysis, presentations, or documentation without regenerating them.

* **Syntax (`plt.savefig()`):**
    * **Basic:** `plt.savefig('filename.png')`
    * **With Options:**
        ```python
        plt.savefig(
            'my_plot.pdf',       # Filename (including path and extension)
            format='pdf',        # Explicitly set format (e.g., 'png', 'pdf', 'svg', 'jpg')
            dpi=300,             # Dots per inch (resolution) for raster formats (PNG, JPG)
            bbox_inches='tight', # Adjusts plot area to fit tightly around content, preventing cut-off labels
            transparent=False,   # Make background transparent
            pad_inches=0.1,      # Padding around the saved plot when bbox_inches='tight'
            quality=95           # Quality for JPEG format (0-100)
        )
        ```
    * `fname` (str or file-like object): The name of the file to save to, including the desired extension (e.g., `'my_plot.png'`). Can also include a path (e.g., `'output/my_plot.png'`).
    * `format` (str, optional): The file format to save the figure. If not specified, it's inferred from the filename extension. Common options include:
        * `'png'` (Portable Network Graphics - Raster)
        * `'pdf'` (Portable Document Format - Vector)
        * `'svg'` (Scalable Vector Graphics - Vector)
        * `'jpg'` / `'jpeg'` (Joint Photographic Experts Group - Raster)
        * `'eps'` (Encapsulated PostScript - Vector)
    * `dpi` (float or 'figure', default: `None`): The resolution in dots per inch. Only applies to raster formats (like PNG, JPG). `'figure'` uses the figure's dpi. Higher DPI means better quality for raster images.
    * `bbox_inches` (str or `Bbox`, default: `'tight'`): How to save the bounding box.
        * `'tight'`: Automatically tightens the bounding box of the figure to include all text, titles, etc., preventing labels from being cut off.
        * `None` or `'standard'`: Uses the figure's current bounding box.
    * `transparent` (bool, default: `False`): If `True`, the background of the plot will be transparent.
    * `pad_inches` (float, default: 0.1): Amount of padding around the figure when `bbox_inches` is `'tight'`.
    * `quality` (int, default: 95): For JPEG images, sets the quality (0-100). Higher is better quality but larger file size.

---

````python
import matplotlib.pyplot as plt
````

````python
# savefig('filename.extension', dpi(dots per inch), bbox_inches = 'tight)

x = [1,2,3,4]

y = [10,20,15,25]

# CREATE PLOT

plt.plot(x,y, color = 'red', marker = 'o')

plt.title("Simple Line Plot")

plt.xlabel("X Axis")

plt.ylabel("Y Axis")

plt.savefig('line_plot.png', dpi = 300, bbox_inches = 'tight')

# dpi = 300 BEST FOR PUBLICATIONS

plt.show()
````
